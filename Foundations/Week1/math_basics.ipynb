# Vectors

```python
import numpy as np

# Define two vectors
vector_a = np.array([1, 2, 3])
vector_b = np.array([4, 5, 6])

print("Vector A:", vector_a)
print("Vector B:", vector_b)
```

```python
import matplotlib.pyplot as plt

# Plot vector_a in the Cartesian plane (2D)
plt.figure(figsize=(3, 3))
plt.quiver(0, 0, vector_a[0], vector_a[1], angles='xy', scale_units='xy', scale=1, color='r')
plt.xlim(0, max(vector_a[0], 1) + 1)
plt.ylim(0, max(vector_a[1], 1) + 1)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Representation of vector_a in 2D Cartesian Plane')
plt.grid()
plt.show()
```

## Vectors - Addition

```python
# Perform vector addition
vector_sum = vector_a + vector_b

print("Sum:", vector_sum)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define vectors
vector_a = np.array([2, 1])
vector_b = np.array([1, 2])
vector_sum = vector_a + vector_b

plt.figure(figsize=(4, 4))
origin = [0], [0]  # origin for quiver

# Plot vector_a
plt.quiver(*origin, vector_a[0], vector_a[1], angles='xy', scale_units='xy', scale=1, color='r', label='vector_a')

# Plot vector_b
plt.quiver(vector_a[0], vector_b[0], vector_b[0], vector_b[1], angles='xy', scale_units='xy', scale=1, color='b', label='vector_b')

# Plot their sum
plt.quiver(*origin, vector_sum[0], vector_sum[1], angles='xy', scale_units='xy', scale=1, color='g', label='a + b')

# Axes setup
max_x = max(vector_a[0], vector_b[0], vector_sum[0]) + 1
max_y = max(vector_a[1], vector_b[1], vector_sum[1]) + 1
plt.xlim(0, max_x)
plt.ylim(0, max_y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Combination of Two 2D Vectors')
plt.grid()
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

## Vectors - Multiplication by Scalar

```python
scalar = 3
result = scalar * vector_a
print("Scalar multiplied by Vector A:", result)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define original vector
v = np.array([2, 1])
scalars = [0.5, 1, 2, -1]
colors = ['orange', 'red', 'green', 'blue']
labels = ['0.5  v', '1  v', '2  v', '-1  v']

plt.figure(figsize=(3, 3))
origin = [0], [0]

# Plot scalar-multiplied vectors as arrows
for scalar, color, label in zip(scalars, colors, labels):
    scaled_v = scalar * v
    plt.quiver(*origin, scaled_v[0], scaled_v[1], angles='xy', scale_units='xy', scale=1, color=color, label=label)

# Plot original vector as dashed line for reference (using plot instead of quiver)
plt.plot([0, v[0]], [0, v[1]], 'k--', label='Original v (ref)')

# Axis setup
max_val = max(abs(s * v[0]) for s in scalars) + 1
plt.xlim(-max_val, max_val)
plt.ylim(-max_val, max_val)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scalar Multiplication of a Vector in 2D')
plt.grid()
plt.gca().set_aspect('equal', adjustable='box')
plt.legend(loc='lower right', fontsize='small')
plt.show()
```
